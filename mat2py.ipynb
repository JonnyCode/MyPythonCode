Import .mat files into Python with:
```Python
scipy.io.loadmat('filename.mat')
```
Output: dictionary with variable names as keys, and variable values as values.

````python
import scipy.io as io
import numpy as np
mat = io.loadmat('BWDb1.mat')
print(mat)
````

Ignore keys with double underscores at the beginning and end like `__this__`. In the above case, there are 3 variables in test.mat, one is 'spikes' from `datarun.spikes`. The other 2 are matrices with random numbers generated by `randn` in MATLAB.

``` MATLAB
spikes = datarun.spikes;
matrix2 = randn(3, 3);
matrix3 = randn(3, 3, 3);
```


````python
print('matrix2:', mat['matrix2'])
print('row 1 column 1 of matrix2:', mat['matrix2'][0,0])
````

````python
print('matrix3:', mat['matrix3'])
print('matrix3[0,0,0]:', mat['matrix3'][0,0,0])
````

````python
print('Spikes of the first cell: ', mat['spikes'][0,0])
print('1st spike of the first cell: ', mat['spikes'][0,0][0,0])

print('Spikes of the second cell: ', mat['spikes'][1,0])
print('2nd spike of the second cell: ', mat['spikes'][1,0][1,0])

print(type(mat['spikes']))
````
